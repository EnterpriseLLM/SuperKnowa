```python
def get_all_files(folder_name):
    # Change the directory
    os.chdir(folder_name)
    # iterate through all file
    file_path_list =[]
    for file in os.listdir():
        file_path = f"{folder_name}/{file}"
        file_path_list.append(file_path)
    return file_path_list
        
    
```

```python
from nltk.tokenize import word_tokenize

def count_tokens_in_file(file_path):
    with open(file_path, 'r', encoding='latin1') as file:
        text = file.read()

    tokens = word_tokenize(text)  # Tokenize the text using NLTK

    return len(tokens)
```

### IBM DOCS 

```python
file_list = get_all_files(ibm_docs_dir)
```

```python
token_page_count_list =[]
total_pages =0
total_count =0
for file in file_list:
    print(file)
    try:
        token_count = count_tokens_in_file(file)
        page_count =  math.ceil(token_count/500)
        total_pages = total_pages+page_count
        total_count = total_count+token_count
        token_page_count_list.append({"File Name":file,"Token Count":token_count,"Page":page_count})
        print(f"Number of tokens in the file: {token_count}")
    except:
        continue
        
```

```python
print("IBM DOCS Total Pages ===",total_pages)
print("IBM DOCS Total Tokens Count ===",total_count)
```

## Red Books PDF

```python
def readdata_frompdf(file_name):
    pdfFileObj = open(file_name, 'rb')
    pdfReader = PyPDF2.PdfReader(pdfFileObj)
    content=''
    for i in range(len(pdfReader.pages)):
        pageObj = pdfReader.pages[i]
        content =content+" "+pageObj.extract_text()
    
    pdfFileObj.close()
    tokens = word_tokenize(content)  # Tokenize the text using NLTK
    return len(tokens)
```
